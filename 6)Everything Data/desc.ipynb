<h3>Lakehouse </h3>
    <img src="https://content.altexsoft.com/media/2021/11/the-two-tier-architecture-with-a-data-lake-and-dat.png.webp" width="500"/>
    <p align="left"><small ><i>Courtesy: altexsoft </i></small></p>
<p>
Data Lakehouse combines the capabilities of both data warehouses and data lakes. A company's analytics team may contain a data analyst,  data scientist, BI engineer etc. Data analysts and BI engineers mostly work with data in tabular format with predefined schemas, unlike data scientists who can work on tabular data, text data, image data, and sound data. Data Lakehouse gives the capability to store several formats which enables data science, data analysis, and BI task in a single location.

<h3>Data Mart</h3>
    <img src="http://mbenhaddou.com/wp-content/uploads/2020/01/datamarts.png" width="500"/>
    <p align="left"><small ><i>Courtesy: mbenhaddou </i></small></p>
<p>
A Data Mart is a subset of a larger data warehouse that is designed to serve a specific department or business function. It contains a subset of the data from the data warehouse that is relevant to that specific department.

<h3>Data Catalog</h3>
    <img src="https://www.tibco.com/sites/tibco/files/media_entity/2021-05/data-catalog-diagram.svg" width="500"/>
    <p align="left"><small ><i>Courtesy: Tibco </i></small></p>
<p>
A Data Catalog is a centralized repository of metadata, or data about data, that describes the data assets available in an organization. It can include information about data sources, relationships between data elements, and the structure of data.

</p>

<h3>Data Quality</h3>

<p>
